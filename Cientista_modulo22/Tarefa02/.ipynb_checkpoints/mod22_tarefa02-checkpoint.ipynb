````python
from pandas_datareader import data as web
import pandas as pd
import plotly.express as px
import datetime as dt
import yfinance as yf
from plotly import graph_objects
from plotly.graph_objs import Layout
````

#### 1. Carregue os dados da tarefa anterior

````python
yf.pdr_override()
ativos = ['PETR4.SA','VALE3.SA','MGLU3.SA']
dt_ini = '2021-01-01'

df = web.get_data_yahoo(ativos, start=dt_ini)
df.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
#Media Movel
df_movel_fechamento = df['Adj Close']
df_movel_fechamento = df_movel_fechamento.rolling(90).mean()
df_movel_fechamento_vertical = df_movel_fechamento.stack().reset_index()
df_movel_fechamento_vertical.columns = ['data','ativo','media_movel']
df_movel_fechamento_vertical
````

````python
#Desvio Padrão
df_desvio_fechamento = df['Adj Close']
df_desvio_fechamento = df_desvio_fechamento.rolling(90).std()
df_desvio_fechamento_vertical = df_desvio_fechamento.stack().reset_index()
df_desvio_fechamento_vertical.columns = ['data','ativo','desvio_padrao']
df_desvio_fechamento_vertical
````

````python
#Plotagem Media Movel
fig = px.line(df_movel_fechamento_vertical, x='data', y='media_movel', color='ativo', template='plotly_white')
fig.show()
````

````python
#Plotagem Desvio Padrão
fig = px.line(df_desvio_fechamento_vertical, x='data', y='desvio_padrao', color='ativo', template='plotly_white')
fig.show()
````

## A relação entre os dois gráficos da para ser notado nos ativos MGLU3.SA e PETR4.SA.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df.columns
````

````python
lista_indicadores = ['High', 'Close', 'Low', 'Open']
ativo = ['PETR4.SA']*len(lista_indicadores)
colunas = list(zip(lista_indicadores, ativo))

petr4 = df.loc['07/22/2023':,colunas]
petr4.columns = lista_indicadores
petr4.head()
````

````python
graph = {
    'x' : petr4.index,
    'open' : petr4.Open,
    'close' : petr4.Close,
    'high' : petr4.High,
    'low' : petr4.Low,
    'type' : 'candlestick',
    'name' : 'PETR4.SA',
    'showlegend' : True
}
````

````python
 layout = Layout(
     paper_bgcolor='rgba(0,0,0,0)',
     plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data=[graph],
    layout_title="Índice PETR4",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
