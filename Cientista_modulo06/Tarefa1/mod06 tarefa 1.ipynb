```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'sinasc_RO_2019.csv'

sinasc = pd.read_csv(url)
```

```python
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10,3))
fig,  axes = plt.subplots(1,2)

sns.countplot(ax = axes[0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[1], x='APGAR5', data=sinasc)
```

```python
plt.rc('figure', figsize=(5,5))
sns.countplot(x='SEXO', data=sinasc)
```

```python
fig, ax = plt.subplots()

gr_gestacao=sns.countplot(x='GESTACAO', data=sinasc)

gr_gestacao.set_xticklabels(gr_gestacao.get_xticklabels(), rotation=45, horizontalalignment='right')
```

```python
sns.countplot(x='ESTCIVMAE', data=sinasc)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)

gr1 = sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
gr2 = sns.countplot(ax = axes[1, 0], x='CONSULTAS', data=sinasc)
gr3 = sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=sinasc)
gr4 = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc)

gr4.set_xticklabels(gr4.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
