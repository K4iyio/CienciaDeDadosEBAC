# Regressão 01 - tarefa 02 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

%matplotlib inline
````

````python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
reg = smf.ols('tip_pct ~ net_bill', data=tips).fit()
print(reg.summary())

sns.regplot(
    data=tips[tips['tip_pct']<.5],
    x='net_bill',
    y='tip_pct')

plt.xlim(left=0)
plt.ylim(bottom=0)
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x = 'net_bill', y = 'res', data = tips[tips['tip_pct']<.5], alpha = .75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x = 'net_bill', y = 'tip_pct', data = tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.power(net_bill,2)', data = tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips[tips['tip_pct']<.5], alpha=.75)
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg = smf.ols('np.log(tip_pct) ~ net_bill', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips[tips['tip_pct']<.5], alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

````python
tips['pred_tip_pct'] = np.exp(reg.fittedvalues)

tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())

tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='res', data=tips[tips['tip_pct']<.5], alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

````python
tips['pred_tip_pct'] = np.exp(reg.fittedvalues)

tips[['pred_tip_pct', 'tip_pct']].corr().iloc[0,1]**2
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips[tips['tip_pct']<.5])
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')
````

###  6. qual modelo resultou no melhor ajuste?

 - O modelo que resultou no melhor ajuste foi o modelo com log de x e log de y.
