````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_theme(style="whitegrid")

df = pd.read_csv('previsao_de_renda.csv')
````

````python
df.columns
````

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda', dodge = True, data = df)
````

````python
ax = sns.pointplot(x='posse_de_imovel', y='renda', dodge = False, data = df)
````
