````python
#Importando Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_theme(style="whitegrid")#Tema Seaborn
df = pd.read_csv('previsao_de_renda.csv')#Lendo .CSV
````

````python
print(df.columns) # Analisando Colunas
df.head()
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref']) #Transformado ['data_ref'] em Datetime
````

````python
df.dtypes
````

````python
#Plotando #x=data_ref hue=sexo
ax = sns.countplot(x='data_ref', hue='sexo', data=df)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
